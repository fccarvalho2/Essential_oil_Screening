## Procurando pela maior Afinidade

```python
dfVina = pd.read_csv('dfVina.csv')
dfVina.head(50)
```

```python
dfVa= dfVina.groupby(['Ligand-ID'], sort=False)['Vina'].min()
idx1 = dfVina.groupby(['Ligand-ID'])['Vina'].transform(min) == dfVina['Vina']
```

```python
Vina = dfVina[idx1]
Vina.to_csv("Vina.csv")
```

```python
dfConvex = pd.read_csv('dfConvex.csv')
dfConvex.head(50)
```

```python
dfCo= dfConvex.groupby(['Ligand-ID'], sort=False)['Convex'].max()
idx2 = dfConvex.groupby(['Ligand-ID'])['Convex'].transform(max) == dfConvex['Convex']
```

```python
Convex = dfConvex[idx2]
Convex
```

```python
Convex.to_csv("Convex.csv")
```

## Unindo as tabelas

```python
from functools import reduce
```

```python
df1 = pd.read_csv('Vina.csv')
df2 = pd.read_csv('Convex.csv')
df3 = pd.read_csv('Vinardo.csv')
```

```python
dfs = [df1, df2, df3]
```

```python
df_final = reduce(lambda left,right: pd.merge(left,right,on='Ligand-ID'), dfs)
```

```python
df_final.to_csv("df_final.csv")
```
