## Calculando as Médias

```python
df = pd.read_csv('JHBPSumario.csv')
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
df
```

```python
rdf = df.groupby('Ligand').mean()
rdf
```

```python
rdf.to_csv('JHBPR.csv')
```
