```python
import os
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
# Docking script using directories as reference
config = "config.txt" #ok
recdir = 'receptores/' #ok
ligdir = 'controles-positivos/' #ok
resultsdir = '' #ok
logsdir = '' #ok
resultsprefix = '' #ok
logsprefix = '' # ok

# preparando os receptores e convertendo de pdb para pdbqt (Loops)
for r in os.listdir(recdir):
    rec_name = recdir + r
    rec_pdbqt = 'PDBQT/Receptores/' + r[:-4] + '.pdbqt'
    prepare = 'mgltools_x86_64Linux2_1.5.7rc1/MGLToolsPckgs/AutoDockTools/Utilities24/prepare_'
    pythonsh = 'mgltools_x86_64Linux2_1.5.7rc1/bin/pythonsh'
    os.system("mgltools_x86_64Linux2_1.5.7rc1/bin/pythonsh " + prepare + "receptor4.py -r " + rec_name + " -o " + rec_pdbqt) #converte o arquivo do receptor para formato pdbqt
    for l in os.listdir(ligdir):
        lig_name = ligdir + l
        lig_pdbqt = 'PDBQT/Ligantes/' + l[:-5] + '.pdbqt'
        os.system("mgltools_x86_64Linux2_1.5.7rc1/bin/pythonsh "+ prepare + "ligand4.py -l "+ lig_name + " -o " + lig_pdbqt) #converte o arquivo do ligante para formato pdbqt
        print(' '.join(['\n\nStarting docking for',l[:-5], 'on', r[:-4],':\n']))
        os.system("vina --receptor "+ rec_pdbqt + " --ligand "+ lig_pdbqt + " --out Resultados/out_"+l[:-5]+"_"+r[:-4]+".pdbqt --log Resultados/Logs/result_"+l[:-5]+"_"+r[:-4]+".log --config config.txt") 
```

```python
def read_file(filename):
    for line in open(filename):
        yield line  

logsdir = 'Resultados/Logs/'
# Optional: Allow user to define how many of the results to use
data = {"Ligand":[], "Receptor":[], "Affinity1":[], "Affinity2":[], "Affinity3":[]}
checagem = []
for filename in os.listdir(logsdir):
    #filename = "result" + name + "ache1.log"
    full_filename = logsdir + filename
    ligname = filename.split('_')[1]
    recname = filename.split('_')[2][:-4]
    data["Ligand"].append(ligname)
    data["Receptor"].append(recname)
    for line in read_file(full_filename):
        if len(line)> 3:
            if line[3] == "1":
                checagem.append(ligname)
                l = line.split("     ")
                data["Affinity1"].append(float(l[1]))
                
            elif line[3] == "2":
                l = line.split("     ")
                data["Affinity2"].append(float(l[1]))
                
            elif line[3] == "3":
                l = line.split("     ")
                data["Affinity3"].append(float(l[1]))
            
#for t in names:
#    if t not in checagem:
#        print(t)
df = pd.DataFrame(data)
df = df.set_index("Ligand", drop=True)
df.to_csv("Sumario_VINA.csv")
df
```
