```python
import pandas as pd
import numpy as np
df1 = pd.read_csv('controleVina.csv')
df2 = pd.read_csv('controleConvex.csv')
```

```python
# Método 1 - Ver se os dois quadros de dados são iguais
array1 = np.array(df1)
array2 = np.array(df2)

df_CSV_1 = pd.DataFrame(array1, columns=['Ligand','Receptor','Affinity1','Affinity2','Affinity3'])
df_CSV_2 = pd.DataFrame(array2, columns=['Ligand','Receptor','Affinity1','Affinity2','Affinity3'])

df_CSV_1.index += 1
df_CSV_2.index += 1

df_CSV_1.index += 1 
df_CSV_2.index += 1

print(df_CSV_1.eq(df_CSV_2).to_string(index=True))
```

```python
a = df1[df1.eq(df2).all(axis=1) == False]

a.index += 1 ===>This resets the index to start at 1, not 0, which helps with the output when trying to understand the differences. 

print(a.to_string(index=False))
```
