```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')
```

```python
rankAChE = pd.read_excel('./input/rank_AChE.xlsx')
rankEcNR = pd.read_excel('./input/rank_EcNR.xlsx')
rankJHBP = pd.read_excel('./input/rank_JHBP.xlsx')
rankMET = pd.read_excel('./input/rank_MET.xlsx')
rankGABAaRago = pd.read_excel('./input/rank_GABAaRagonist.xlsx')
rankGABAaRantg = pd.read_excel('./input/rank_GABAaRantagonist.xlsx')
rankOctpRago = pd.read_excel('./input/rank_OctpR_agonist.xlsx')
rankOctpRantgo = pd.read_excel('./input/rank_OctpR_antagonist.xlsx')
```

```python
rankAChE = rankAChE[['Compound','rank_Medio']]
rankEcNR = rankEcNR[['Compound','rank_Medio']]
rankJHBP = rankJHBP[['Compound','rank_Medio']]
rankMET = rankMET[['Compound','rank_Medio']]
rankGABAaRago = rankGABAaRago[['Compound','rank_Medio']]
rankGABAaRantg = rankGABAaRantg[['Compound','rank_Medio']]
rankOctpRago = rankOctpRago[['Compound','rank_Medio']]
rankOctpRantgo = rankOctpRantgo[['Compound','rank_Medio']]
```

```python
rankAChE
```

```python
def composto(s):
    return s.split('-')[0]
rankAChE['Compound'] = rankAChE['Compound'].apply(composto)
rankEcNR['Compound'] = rankEcNR['Compound'].apply(composto)
rankJHBP['Compound'] = rankJHBP['Compound'].apply(composto)
rankMET['Compound'] = rankMET['Compound'].apply(composto)
rankGABAaRago['Compound'] = rankGABAaRago['Compound'].apply(composto)
rankGABAaRantg['Compound'] = rankGABAaRantg['Compound'].apply(composto)
rankOctpRago['Compound'] = rankOctpRago['Compound'].apply(composto)
rankOctpRantgo['Compound'] = rankOctpRantgo['Compound'].apply(composto)
```

```python
distEuclidiana = pd.read_csv('./input/EPNMCS_vs_Matrix1.csv')
distEuclidiana
```

```python
mAChE = pd.merge(distEuclidiana, rankAChE, how = 'inner', on = 'Compound')
mEcNR = pd.merge(distEuclidiana, rankAChE, how = 'inner', on = 'Compound')
mJHBP = pd.merge(distEuclidiana, rankAChE, how = 'inner', on = 'Compound')
mMET = pd.merge(distEuclidiana, rankAChE, how = 'inner', on = 'Compound')
mGABAaRAgo = pd.merge(distEuclidiana, rankAChE, how = 'inner', on = 'Compound')
mGABAaRAnt = pd.merge(distEuclidiana, rankAChE, how = 'inner', on = 'Compound')
mOctpRAgo = pd.merge(distEuclidiana, rankAChE, how = 'inner', on = 'Compound')
mOctpRAnt = pd.merge(distEuclidiana, rankAChE, how = 'inner', on = 'Compound')
```

```python
corrAChE = mAChE[['AChE','rank_Medio']]
corrEcNR = mEcNR[['EcNR','rank_Medio']]
corrJHBP = mJHBP[['JHBP','rank_Medio']]
corrMET = mMET[['MET','rank_Medio']]
corrGABAaRAgo = mGABAaRAgo[['GABAaR-Ago','rank_Medio']]
corrGABAaRAnt = mGABAaRAnt[['GABAaR-Ant','rank_Medio']]
corrOctpRAgo = mOctpRAgo[['OctpR-Ago','rank_Medio']]
corrOctpRAnt = mOctpRAnt[['OctpR-Ant','rank_Medio']]
```

```python
corrAChE.to_csv('./outputcorrAChE.csv')
corrEcNR.to_csv('corrEcNR.csv')
corrJHBP.to_csv('corrJHBP.csv')
corrMET.to_csv('corrMET.csv')
corrGABAaRAgo.to_csv('corrGABAaRAgo.csv')
corrGABAaRAnt.to_csv('corrGABAaRAnt.csv')
corrOctpRAgo.to_csv('corrOctpRAgo.csv')
corrOctpRAnt.to_csv('corrOctpRAnt.csv')
```
