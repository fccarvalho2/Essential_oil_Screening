```python
# library
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
matrizfracaocomponente = pd.read_csv('./input/fracao_componentes.csv', index_col='Compound')
matrizfracaocomponente = matrizfracaocomponente.transpose()
matrizfracaocomponente
```

```python
dfigualmenteprovaveis = pd.read_csv('./input/matrizprovaveis.csv', index_col='Unnamed: 0')
dfigualmenteprovaveis
```

```python
MatrizAcaoProvavelOleos = matrizfracaocomponente@dfigualmenteprovaveis
MatrizAcaoProvavelOleos
```

### Visualizando a matrix de probabilidade de um oleo afetar um alvo molecular de inseto

```python
#Ajust figure size
fig, ax = plt.subplots(figsize=(14, 9))
#df = round(matrizFinal1, 3)
#cmap = sns.diverging_palette(220, 20, as_cmap=True)
p1 = sns.heatmap(MatrizAcaoProvavelOleos,cmap='Blues',
                 linewidths=0.75,
                 linecolor="lightgray",
                 annot = True, annot_kws = {"fontsize":23, 'rotation': 0}
                )
p1.set_yticklabels(p1.get_yticklabels(), rotation = 0, fontsize = 18, style='italic')
p1.set_xticklabels(p1.get_xticklabels(), rotation = 90, fontsize = 18)
#plt.title('Matrix Cartesinan Proximity EO per Protein Target - Matrix 1', fontsize = 10, fontweight = 'bold')
plt.savefig("FiguraFinal-OleosEssencialVrsAlvo.png", dpi = 150, facecolor = "white")

plt.show()
```
