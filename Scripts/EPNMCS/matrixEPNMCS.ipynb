```python
# library to import
import pandas as pd
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plan = pd.read_csv("./input/descritores.csv")
plan = plan.drop_duplicates("Ligante")
plan = plan.fillna(plan.median())
plan
```

```python
tabelahits = plan[['Ligante',
               'Molecular Weight',
               'Net Charge',
               'Net Charge Module',
               'Dipole Moment',
               'Molar Refractivity',
               'Molar Volume',
               'Parachor',
               'Index of Refraction',
               'Surface Tension',
               'Density',
               'Polarizability',
               'RDBE',
               'miLogP',
               'tPSA',
               'natoms',
               'Donors',
               'Acceptors',
               'nviolations',
               'Rotatable Bonds',
               'LogP','Length',
               'Max Length',
               'Extensibility']]
medias = ["Media"]
desvios = ["Desvio"]
for c in tabelahits.columns:
    if c != "Ligante":
        medias.append(tabelahits[c].mean())
        desvios.append(tabelahits[c].std())
df = pd.DataFrame([medias, desvios], columns=tabelahits.columns)
tabelahits = pd.concat([tabelahits,df])
tabelahits
```

```python
## Calcula a media de cada descritor para cada cluster na tabela xlcluster.xlsx
ddfcluster = pd.DataFrame(columns=tabelahits.columns)
xl = pd.ExcelFile('./input/xlcluster.xlsx')
for nome in xl.sheet_names:
    novo_data = pd.read_excel('./input/xlcluster.xlsx', sheet_name=nome)
    try:
        rec_data = tabelahits.loc[tabelahits['Ligante'].isin(novo_data.Ligante)]
    except:
        rec_data = tabelahits.loc[tabelahits['Ligante'].isin(novo_data.ligante)]
    m1 = [nome]
    m1.extend(rec_data.mean())
    media_rec = pd.DataFrame(m1).transpose()
    media_rec.columns =tabelahits.columns
    ddfcluster = pd.concat([ddfcluster, media_rec])
ddfcluster.rename(columns= {"Ligante":"Cluster"}, inplace=True)
ddfcluster.to_excel('./output/MatrizReferenciaCluster.xlsx')
ddfcluster
```

```python
#Calcula o Z-score de cada descritor para cada cluster
zscoresCluster = ddfcluster.copy(deep=True)
for coluna in zscoresCluster.columns[1:]:
    if coluna == "Net Charge":
        continue
    lista = []
    media = tabelahits[tabelahits["Ligante"]=="Media"][coluna]
    dp = tabelahits[tabelahits["Ligante"]=="Desvio"][coluna]
    for r in zscoresCluster.Cluster:
        valor = ddfcluster[ddfcluster["Cluster"]==r][coluna]
        z = (valor[0] - media[0])/list(dp)[0]
        lista.append(z)
    zscoresCluster[coluna] = lista
zscoresCluster.to_excel('./output/ZScorePorAlvoCalculadoCluster.xlsx')
zscoresCluster
```

```python
## Aqui ele vai calcular a distancia euclidiana do cluster versus o cluster
matrizClusters = pd.DataFrame(index = zscoresCluster.Cluster.values,columns = zscoresCluster.Cluster.values)
for coluna in zscoresCluster.Cluster.values:
    lista = []
    for linha in zscoresCluster.Cluster.values:
        a = zscoresCluster.loc[zscoresCluster["Cluster"]==linha].values[0][1:]
        b = zscoresCluster.loc[zscoresCluster["Cluster"]==coluna].values[0][1:]
        dist = np.linalg.norm(a-b)
        lista.append(dist)
    matrizClusters[coluna] = lista
matrizClusters.to_csv("./output/matriz1-Cluster.csv")
```

```python
## Montando a tabela com Ligantes que compoem os Oleos Essenciais
oleos = pd.read_csv('./input/descritoresOleos.csv')
oleos = oleos.drop_duplicates("Ligante")
oleos = oleos.fillna(plan.median())
oleos = oleos[['Ligante',
               'Molecular Weight',
               'Net Charge',
               'Net Charge Module',
               'Dipole Moment',
               'Molar Refractivity',
               'Molar Volume',
               'Parachor',
               'Index of Refraction',
               'Surface Tension',
               'Density',
               'Polarizability',
               'RDBE',
               'miLogP',
               'tPSA',
               'natoms',
               'Donors',
               'Acceptors',
               'nviolations',
               'Rotatable Bonds',
               'LogP','Length',
               'Max Length',
               'Extensibility']]
medias = ["Media"]
desvios = ["Desvio"]
for c in oleos.columns:
    if c != "Ligante":
        medias.append(oleos[c].mean())
        desvios.append(oleos[c].std())
df = pd.DataFrame([medias, desvios], columns=oleos.columns)
oleos = pd.concat([oleos,df])
oleos
```

```python
## Z-Score dos descitores em relação a media em cda ligante que compoe o Oleo Essencial
z_ligantesOleos = oleos.copy(deep="True")
for coluna in z_ligantesOleos.columns[1:]:
    if coluna == "Net Charge":
        continue
    lista = []
    media = oleos[oleos["Ligante"]=="Media"][coluna]
    dp = oleos[oleos["Ligante"]=="Desvio"][coluna]
    for l in z_ligantesOleos["Ligante"]:
        valor = z_ligantesOleos[z_ligantesOleos["Ligante"]==l][coluna]
        z = (list(valor)[0] - list(media)[0])/list(dp)[0]
        lista.append(z)
    z_ligantesOleos[coluna] = lista
z_ligantesOleos
```

```python
## Calculo da distancia euclidiana dos Z_scores de cada ligante em relação ao Z_score de cada alvo.
cluster_ligante = pd.DataFrame(index = z_ligantesOleos.Ligante.values,columns = zscoresCluster.Cluster.values)
for coluna in zscoresCluster.Cluster.values:
    lista = []
    for linha in z_ligantesOleos.Ligante.values:
        a = zscoresCluster.loc[zscoresCluster["Cluster"]==coluna].values[0][1:]
        b = z_ligantesOleos.loc[z_ligantesOleos["Ligante"]==linha].values[0][1:]
        dist = np.linalg.norm(a-b)
        lista.append(dist)
    cluster_ligante[coluna] = lista
cluster_ligante = cluster_ligante.drop(["Media", "Desvio"])
#alvo_ligante = receptor_ligante.transpose()

cluster_ligante
```
